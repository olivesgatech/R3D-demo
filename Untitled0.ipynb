```python
# 6. calculating effective rank

log_idx = list(range(33))
obs_perc_list = [round(p, 1) for p in np.arange(0.1, 1.0, 0.1)]

def rankMe_Score(Z):
    svd = torch.linalg.svdvals(Z)
    svd = svd[svd > 1e-12]
    p = svd / (svd.sum() + 1e-7)
    rankme_score = -torch.sum(p * torch.log(p + 1e-7))
    return torch.exp(rankme_score).item()
```

```python
# 7. RGB/depth before fusion
datasets = 'darai_depth'

raw_rgb_eranks = []
raw_depth_eranks = []

for obs_perc in obs_perc_list:
    sum_raw_rgb = 0.0
    sum_raw_depth = 0.0

    for idx in log_idx:
        raw_rgb = torch.load(f"pt_{datasets}/rgb_raw/rgb_raw_{idx}_{obs_perc}.pt")[0]
        raw_depth = torch.load(f"pt_{datasets}/depth_raw/depth_raw_{idx}_{obs_perc}.pt")[0]

        sum_raw_rgb += rankMe_Score(raw_rgb)
        sum_raw_depth += rankMe_Score(raw_depth)

    raw_rgb_eranks.append(sum_raw_rgb / len(log_idx))
    raw_depth_eranks.append(sum_raw_depth / len(log_idx))

plt.figure(figsize=(8, 5))
plt.plot(obs_perc_list, raw_rgb_eranks, label='Raw RGB Effective Rank', marker='o')
plt.plot(obs_perc_list, raw_depth_eranks, label='Raw Depth Effective Rank', marker='x')
plt.xlabel('Observation Percentage')
plt.ylabel('Effective Rank')
plt.title('Raw Effective Rank vs. Observation Percentage')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 11. Effective rank gain for both pairs

datasets = 'darai_depth'
rgb_erank_gains = []
depth_erank_gains = []

for obs_perc in obs_perc_list:
    sum_scores = {
        "raw_rgb": 0,
        "exchanged_rgb": 0,
        "raw_depth": 0,
        "exchanged_depth": 0,
    }

    for idx in log_idx:
        raw_rgb = torch.load(f"pt_{datasets}/rgb_raw/rgb_raw_{idx}_{obs_perc}.pt")[0]
        exchanged_rgb = torch.load(f"pt_{datasets}/exchanged_rgb/exchanged_rgb_{idx}_{obs_perc}.pt")[0]
        raw_depth = torch.load(f"pt_{datasets}/depth_raw/depth_raw_{idx}_{obs_perc}.pt")[0]
        exchanged_depth = torch.load(f"pt_{datasets}/exchanged_depth/exchanged_depth_{idx}_{obs_perc}.pt")[0]

        sum_scores["raw_rgb"] += rankMe_Score(raw_rgb)
        sum_scores["exchanged_rgb"] += rankMe_Score(exchanged_rgb)
        sum_scores["raw_depth"] += rankMe_Score(raw_depth)
        sum_scores["exchanged_depth"] += rankMe_Score(exchanged_depth)

    mean_scores = {key: val / len(log_idx) for key, val in sum_scores.items()}
    rgb_erank_gains.append(mean_scores["exchanged_rgb"] - mean_scores["raw_rgb"])
    depth_erank_gains.append(mean_scores["exchanged_depth"] - mean_scores["raw_depth"])

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(obs_perc_list, rgb_erank_gains, label='RGB Effective Rank Gain', marker='o')
plt.plot(obs_perc_list, depth_erank_gains, label='Depth Effective Rank Gain', marker='x')
plt.xlabel('Observation Percentage')
plt.ylabel('Effective Rank Gain')
plt.ylim(0, 4)
plt.title('Effective Rank Gain vs. Observation Percentage')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


datasets = 'darai_erank_text'
rgb_erank_gains = []
depth_erank_gains = []

for obs_perc in obs_perc_list:
    sum_scores = {
        "raw_rgb": 0,
        "exchanged_rgb": 0,
        "raw_depth": 0,
        "exchanged_depth": 0,
    }

    for idx in log_idx:
        raw_rgb = torch.load(f"pt_{datasets}/rgb_raw/rgb_raw_{idx}_{obs_perc}.pt")[0]
        exchanged_rgb = torch.load(f"pt_{datasets}/exchanged_rgb/exchanged_rgb_{idx}_{obs_perc}.pt")[0]
        raw_depth = torch.load(f"pt_{datasets}/depth_raw/depth_raw_{idx}_{obs_perc}.pt")[0]
        exchanged_depth = torch.load(f"pt_{datasets}/exchanged_depth/exchanged_depth_{idx}_{obs_perc}.pt")[0]

        sum_scores["raw_rgb"] += rankMe_Score(raw_rgb)
        sum_scores["exchanged_rgb"] += rankMe_Score(exchanged_rgb)
        sum_scores["raw_depth"] += rankMe_Score(raw_depth)
        sum_scores["exchanged_depth"] += rankMe_Score(exchanged_depth)

    mean_scores = {key: val / len(log_idx) for key, val in sum_scores.items()}
    rgb_erank_gains.append(mean_scores["exchanged_rgb"] - mean_scores["raw_rgb"])
    depth_erank_gains.append(mean_scores["exchanged_depth"] - mean_scores["raw_depth"])

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(obs_perc_list, rgb_erank_gains, label='RGB Effective Rank Gain', marker='o')
plt.plot(obs_perc_list, depth_erank_gains, label='Depth Effective Rank Gain', marker='x')
plt.xlabel('Observation Percentage')
plt.ylabel('Effective Rank Gain')
plt.ylim(0, 4)
plt.title('Effective Rank Gain vs. Observation Percentage')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 12. Calculate harmonic mean

def compute_harmonic_erank_gains(datasets, log_idx):
    rgb_erank_gains = []
    depth_erank_gains = []

    for obs_perc in obs_perc_list:
        sum_scores = {
            "raw_rgb": 0,
            "exchanged_rgb": 0,
            "raw_depth": 0,
            "exchanged_depth": 0,
        }

        for idx in log_idx:
            raw_rgb = torch.load(f"pt_{datasets}/rgb_raw/rgb_raw_{idx}_{obs_perc}.pt")[0]
            exchanged_rgb = torch.load(f"pt_{datasets}/exchanged_rgb/exchanged_rgb_{idx}_{obs_perc}.pt")[0]
            raw_depth = torch.load(f"pt_{datasets}/depth_raw/depth_raw_{idx}_{obs_perc}.pt")[0]
            exchanged_depth = torch.load(f"pt_{datasets}/exchanged_depth/exchanged_depth_{idx}_{obs_perc}.pt")[0]

            sum_scores["raw_rgb"] += rankMe_Score(raw_rgb)
            sum_scores["exchanged_rgb"] += rankMe_Score(exchanged_rgb)
            sum_scores["raw_depth"] += rankMe_Score(raw_depth)
            sum_scores["exchanged_depth"] += rankMe_Score(exchanged_depth)

        mean_scores = {key: val / len(log_idx) for key, val in sum_scores.items()}
        rgb_erank_gains.append(mean_scores["exchanged_rgb"] - mean_scores["raw_rgb"])
        depth_erank_gains.append(mean_scores["exchanged_depth"] - mean_scores["raw_depth"])

    harmonic_erank_gains = []
    for rgb_gain, depth_gain in zip(rgb_erank_gains, depth_erank_gains):
        if rgb_gain <= 0 or depth_gain <= 0:
            harmonic_erank_gains.append(0.0)
        else:
            h = 2 * rgb_gain * depth_gain / (rgb_gain + depth_gain)
            harmonic_erank_gains.append(h)

    return harmonic_erank_gains

dataset_settings = {
    "darai_depth": list(range(33)),
    "darai_erank_text": list(range(33))
}

harmonic_results = {}
for ds_name, log_idx in dataset_settings.items():
    print(f"Computing harmonic gains for {ds_name} ...")
    harmonic_results[ds_name] = compute_harmonic_erank_gains(ds_name, log_idx)

print("Harmonic gains (darai_depth):", harmonic_results["darai_depth"])
print("Harmonic gains (darai_erank_text):", harmonic_results["darai_erank_text"])

plt.figure(figsize=(8, 5))
plt.plot(
    obs_perc_list,
    harmonic_results["darai_depth"],
    label="RGB + Depth",
    marker="o"
)
plt.plot(
    obs_perc_list,
    harmonic_results["darai_erank_text"],
    label="RGB + Text",
    marker="s"
)

plt.xlabel('Observation Percentage')
plt.ylabel('Effective Rank Gain (H-mean)')
plt.ylim(0, 4)
plt.title('Harmonic Mean Effective Rank Gain vs. Observation Percentage')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
